```
def classifier(X , Y, method = 1):
  from sklearn.preprocessing import StandardScaler
  from sklearn.metrics import confusion_matrix
  from sklearn.model_selection import train_test_split
  if method == 4:
    from sklearn.naive_bayes import GaussianNB
    print("Gaussian Navie Bayes")
    cla = GaussianNB()
  # elif method == 5:
  #   print('Decision Tree')
  #   from sklearn.tree import DecisionTreeRegressor
  #   cla =   reg = DecisionTreeRegressor(max_depth=5)
  elif method == 3:
    print('SVC')
    from sklearn.svm import SVC
    cla = SVC()
  elif method == 1:
    print('Logistic Regression')
    from sklearn.linear_model import LogisticRegression
    cla = LogisticRegression(random_state = 0,solver='lbfgs')
  elif method == 2:
    print('K Neighbours')
    from sklearn.neighbors import KNeighborsClassifier
    cla = KNeighborsClassifier(n_neighbors= 7)
  SS= StandardScaler()
  x = SS.fit_transform(X)
  Xtrain, Xtest, Ytrain, Ytest = train_test_split(x, Y, test_size=0.25, random_state = 0)
  cla.fit(Xtrain, Ytrain)
  print('Score: ',cla.score(Xtest, Ytest)*100)
  ypred = cla.predict(Xtest)
  print("Confusion Matrix: ", confusion_matrix(Ytest, ypred))
  px.scatter_3d(data_frame= bank, x= 'duration', y= 'y', z= 'nr_employed', color = 'ypred')
  return cla.score(Xtest, Ytest)*100
```

```

X = bank[list(bank.columns)[:-1]]
Y= bank['y']
result=[]
for i in range(1,5):
  result.append(classifier(X, Y, method =i))
  print()
for i,e in enumerate(result):
  plt.bar(x = i, height = e)
plt.show()
```
